```python
import json
import folium

with open('Ulsan.json',mode='rt',encoding='utf-8') as f:
    geo = json.loads(f.read())
    
#울산에 해당하는 Geojson 자료를 여는 코드입니다.

    
m = folium.Map(
    location=[36,127],
    zoom_start=6
)
    
# 먼저 지도를 생성해주는 코드입니다.

folium.GeoJson(
    geo,
    name='good_municipalities'
).add_to(m)

#위에서 만든 지도에, 울산광역시만 구획을 나누어 입혀주는 코드입니다.

m
```
