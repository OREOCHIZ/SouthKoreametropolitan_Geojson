```python
import json
import folium

with open('Ulsan.json',mode='rt',encoding='utf-8') as f:
    geo = json.loads(f.read())
    
#  Geojson   .

    
m = folium.Map(
    location=[36,127],
    zoom_start=6
)
    
#    .

folium.GeoJson(
    geo,
    name='good_municipalities'
).add_to(m)

#  ,     .

m
```
